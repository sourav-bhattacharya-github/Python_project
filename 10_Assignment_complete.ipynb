```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# constants (make sure they match the model you trained)
NOISE_DIM = 100

# Load pretrained generator
generator = tf.keras.models.load_model("/content/Face-Generator-with-GAN/generator_600.h5", compile=False)
generator.summary()

# Function to show generated samples
def show_generated_images(model, n=16, noise_dim=NOISE_DIM):
    noise = tf.random.normal([n, noise_dim])
    preds = model(noise, training=False)
    preds = (preds + 1) / 2.0

    rows = int(np.sqrt(n))
    cols = int(np.ceil(n / rows))
    fig, axs = plt.subplots(rows, cols, figsize=(cols*2, rows*2))
    axs = axs.flatten()

    for i in range(n):
        axs[i].imshow(preds[i])
        axs[i].axis("off")
    plt.tight_layout()
    plt.show()

# Generate a 4x4 grid of images
show_generated_images(generator, n=16)
```

#**2. Load Gender Vector**

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Load pretrained generator
generator = tf.keras.models.load_model("/content/Face-Generator-with-GAN/generator_600.h5", compile=False)
NOISE_DIM = 100

# Load gender vector (pre-computed and saved as .npy)
gender_vector = np.load("/content/Gender-Style-Transfer/gender_vec.npy")
gender_vector = gender_vector / np.linalg.norm(gender_vector)

# Function to traverse latent space along gender vector
def traverse_gender(generator, base_noise=None, steps=7, scale=2.0):
    if base_noise is None:
        base_noise = tf.random.normal([1, NOISE_DIM])
    base_noise = base_noise.numpy()

    # Move along gender direction
    alphas = np.linspace(-scale, scale, steps)
    latent_points = [base_noise + a * gender_vector for a in alphas]
    latent_points = tf.convert_to_tensor(np.vstack(latent_points), dtype=tf.float32)

    # Generate images
    images = generator(latent_points, training=False)
    images = (images + 1) / 2.0

    # Plot
    fig, axes = plt.subplots(1, steps, figsize=(steps*2, 2))
    for i, ax in enumerate(axes):
        ax.imshow(images[i])
        ax.axis("off")
        ax.set_title(f"{alphas[i]:.1f}")
    plt.show()

# Example: traverse from -2 → +2 along gender vector
traverse_gender(generator, steps=7, scale=2.0)
```

#**3. Generate 10 variations of the image from male to female as per gender vector**

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Load pretrained generator
generator = tf.keras.models.load_model("/content/Face-Generator-with-GAN/generator_700.h5", compile=False)
NOISE_DIM = 100

# Load precomputed gender vector (make sure this file exists)
gender_vector = np.load("/content/Gender-Style-Transfer/gender_vec.npy")
gender_vector = gender_vector / np.linalg.norm(gender_vector)

# Pick a random base latent point
base_noise = tf.random.normal([1, NOISE_DIM]).numpy()

def generate_gender_variations(generator, base_noise, gender_vector, steps=10, scale=2.0):
    """
    Generate images by moving base_noise along gender_vector.
    steps: number of images
    scale: how far to move in both directions (-scale to +scale)
    """
    alphas = np.linspace(-scale, scale, steps)
    latent_points = [base_noise + a * gender_vector for a in alphas]
    latent_points = tf.convert_to_tensor(np.vstack(latent_points), dtype=tf.float32)

    # Generate images
    images = generator(latent_points, training=False)
    images = (images + 1) / 2.0

    # Plot row of images
    fig, axes = plt.subplots(1, steps, figsize=(steps*2, 2))
    for i, ax in enumerate(axes):
        ax.imshow(images[i])
        ax.axis("off")
        ax.set_title(f"{alphas[i]:.1f}")
    plt.show()

# Run: 10 variations male→female
generate_gender_variations(generator, base_noise, gender_vector, steps=10, scale=2.0)
```
